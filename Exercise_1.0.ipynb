## Dataset Information

## Features

Number of Instances: 20640

Number of Attributes: 8 numeric, predictive attributes and the target

Attribute Information:

MedInc - median income in block group

HouseAge - median house age in block group

AveRooms - average number of rooms per household

AveBedrms - average number of bedrooms per household

Population - block group population

AveOccup - average number of household members

Latitude - block group latitude

Longitude - block group longitude

## Target
The target variable is the median house value for California districts, expressed in hundreds of thousands of dollars ($100,000).

```python
from sklearn.datasets import fetch_california_housing
from sklearn.feature_selection import VarianceThreshold
import pandas as pd
```

```python
housing = fetch_california_housing(as_frame=True)
df = pd.concat([housing.data, housing.target], axis=1)
```

```python
df_housing = pd.DataFrame(housing.data, columns=housing.feature_names)
df_housing_target = pd.DataFrame(housing.target, columns=['MedHouseVal'])
```

1. Use any filter method to select the best features

```python
selector = VarianceThreshold(threshold=0.2) # Variance threshold
sel = selector.fit(df_housing)
sel_index = sel.get_support()
df_housing_norm_vt = df_housing.iloc[:, sel_index]
```

```python
df_housing_norm_vt
```

2. Use any wrapper method to select the best features

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE

model_rf = RandomForestRegressor(n_estimators=500, random_state=0, max_depth = 3)
selector = RFE(model_rf, n_features_to_select=4, step=1)

selector = selector.fit(df_housing, df_housing_target.values.ravel())
selector_ind = selector.get_support()
df_housing_rfe = df_housing.iloc[:, selector_ind]
df_housing_rfe.columns
```

3. Use any embedded methood to select the best features

```python
from re import X
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_housing, df_housing_target, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_selection import SelectFromModel

model_sfe = RandomForestRegressor(n_estimators=500, random_state=0, max_depth = 3)
model_sfe.fit(X_train, y_train.values.ravel())

model_sfe.predict(X_test)

sel_sfm = SelectFromModel(model_sfe, prefit=True)
sel_sfm_index = sel_sfm.get_support()
df_housing_sfm = df_housing.iloc[:, sel_sfm_index]
df_housing_sfm.columns
```

comparison

```python
from sklearn.metrics import mean_squared_error, r2_score
```

default

```python
model_default = RandomForestRegressor(n_estimators=500, random_state=0, max_depth = 3)
model_default.fit(X_train, y_train)
```

filter

```python
from sklearn.model_selection import train_test_split

X_train_fr, X_test_fr, y_train_fr, y_test_fr = train_test_split(df_housing_norm_vt, df_housing_target, test_size=0.2, random_state=42)
```

```python
model_filter = RandomForestRegressor(n_estimators=500, random_state=0, max_depth = 3)
model_filter.fit(X_train_fr, y_train_fr)
```


wrapper

```python
from sklearn.model_selection import train_test_split

X_train_rfe, X_test_rfe, y_train_rfe, y_test_rfe = train_test_split(df_housing_rfe, df_housing_target, test_size=0.2, random_state=42)
```

```python
model_wrapper = RandomForestRegressor(n_estimators=500, random_state=0, max_depth = 3)
model_wrapper.fit(X_train_rfe, y_train_rfe)
```

embedded

```python
from sklearn.model_selection import train_test_split

X_train_sfm, X_test_sfm, y_train_sfm, y_test_sfm = train_test_split(df_housing_sfm, df_housing_target, test_size=0.2, random_state=42)
```

```python
model_embedded = RandomForestRegressor(n_estimators=500, random_state=0, max_depth = 3)
model_embedded.fit(X_train_sfm, y_train_sfm)
```

pred

```python
default_preds = model_default.predict(X_test)
fr_preds = model_filter.predict(X_test_fr)
rfe_preds = model_wrapper.predict(X_test_rfe)
sfm_preds = model_embedded.predict(X_test_sfm)
```

```python
default = mean_squared_error(y_test, default_preds)
fr_mse = mean_squared_error(y_test_fr, fr_preds)
rfe_mse = mean_squared_error(y_test_rfe, rfe_preds)
sfm_mse = mean_squared_error(y_test_sfm, sfm_preds)
```

```python
print(f'Default MSE: {default}')
print(f'Filter MSE: {fr_mse}')
print(f'Wrapper MSE: {rfe_mse}')
print(f'Embedded MSE: {sfm_mse}')
```
